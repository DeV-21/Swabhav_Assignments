3. Plot a bar chart showing the total sales ('Total') for each 'Product Line'.


```python
import matplotlib.pyplot as plt

total_sales = df.groupby('Product line')['Total'].sum()
total_sales.plot(kind='bar', figsize=(10,6), color='skyblue')
plt.ylabel('Total Sales')
plt.title('Total Sales by Product Line')
plt.xticks(rotation=45)
plt.show()
```

4. Determine the top 3 'Product Lines' with the highest 'Gross Income'.

5. Analyze patterns in the use of 'Payment' methods across different 'Customer Types'.


```python
payment_patterns = pd.crosstab(df['Customer type'], df['Payment'], normalize='index') * 100
payment_patterns
```

6. Calculate the cumulative sum of 'Gross Income' over time and plot it.


```python
df_sorted = df.sort_values('Date')
df_sorted['Cumulative_Gross_Income'] = df_sorted['gross income'].cumsum()

plt.figure(figsize=(12,6))
plt.plot(df_sorted['Date'], df_sorted['Cumulative_Gross_Income'], color='green')
plt.xlabel('Date')
plt.ylabel('Cumulative Gross Income')
plt.title('Cumulative Gross Income Over Time')
plt.show()
```

7. Perform a time-series analysis to find the monthly trend in sales by grouping transactions by 'Date'.


```python
df['Month'] = df['Date'].dt.to_period('M')
monthly_sales = df.groupby('Month')['Total'].sum()

monthly_sales.plot(kind='line', marker='o', figsize=(10,6))
plt.ylabel('Total Sales')
plt.title('Monthly Sales Trend')
plt.xticks(rotation=45)
plt.show()
```

8. Using the 'Rating' column, analyze if there is a correlation between customer satisfaction and the product line they purchase.


```python
avg_rating_product = df.groupby('Product line')['Rating'].mean().sort_values()

avg_rating_product.plot(kind='bar', color='purple', figsize=(10,6))
plt.ylabel('Average Rating')
plt.title('Average Rating by Product Line')
plt.xticks(rotation=45)
plt.show()
```

9. Remove any outliers from the 'Total' column where the value is greater than 1000 and analyze the impact on the overall dataset.

```python
print("Original shape:", df.shape)
df_clean = df[df['Total'] <= 1000]
print("After removing outliers shape:", df_clean.shape)

mean_before = df['Total'].mean()
mean_after = df_clean['Total'].mean()

print(f"Mean Total before: {mean_before:.2f}")
print(f"Mean Total after: {mean_after:.2f}")
```
